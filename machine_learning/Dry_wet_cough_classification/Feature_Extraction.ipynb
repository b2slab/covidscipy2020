Features from two folder are read and features are extraced and stored as .csv

```python
from pyAudioAnalysis import MidTermFeatures as mF 
[mid_term_features_dry, wav_file_list2, mid_feature_names] =  mF.directory_feature_extraction("./trimmed_cough_data/dry", 0.1,0.1, 0.01, 0.01, compute_beat=False);
[mid_term_features_wet, wav_file_list2, mid_feature_names] =  mF.directory_feature_extraction("./trimmed_cough_data/wet", 0.1,0.1, 0.01, 0.01, compute_beat=False);
```

```python
len(mid_feature_names)
```

```python
import numpy as np
import pandas as pd
label_dry = np.zeros(np.shape(mid_term_features_dry)[0])
np.shape(label_dry)
label_wet = np.ones(np.shape(mid_term_features_wet)[0])
np.shape(label_wet)
features = np.concatenate((mid_term_features_dry, mid_term_features_wet))
labels = np.concatenate((label_dry, label_wet))
mid_feature_names = np.array(mid_feature_names)

df= pd.DataFrame(features, columns = mid_feature_names)
df["align"] = pd.Series(labels)


df.to_csv(r'./trimmed_cough_data/features.csv', index = False, header=True)
```
